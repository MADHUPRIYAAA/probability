```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/student_data_250.csv')

# If you have data in other format change the read function accordingly.
# For excel: df = pd.read_excel('your_data.xlsx')
# For json: df = pd.read_json('your_data.json')

# Assume the CSV has a column named 'CGPA'
cgpa_scores = df['GPA']

# Define the number of bins
bins = np.arange(2.0, 4.1, 0.25)  # Bins from 2.0 to 4.0 in intervals of 0.2

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(cgpa_scores, bins=bins, kde=True, color='skyblue', edgecolor='black')
plt.xlabel('CGPA')
plt.ylabel('Number of Students')

plt.title('Histogram of CGPA for 250 Students')
plt.xticks(bins)

plt.grid(axis='y', linestyle='--', alpha=0.7)

# Analyze distribution shape
mean_cgpa = np.mean(cgpa_scores)
median_cgpa = np.median(cgpa_scores)

skewness = df['GPA'].skew()
print(f"Mean GPA: {mean_cgpa:.2f}, Median CGPA: {median_cgpa:.2f}, Skewness: {skewness:.2f}")
plt.show()
```
