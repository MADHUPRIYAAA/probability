```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Set random seed for reproducibility
np.random.seed(42)
N = 1000  # Number of batches
max_defects = 8  # Defective items range from 0 to 8
# Generate a random dataset of defective items per batch
data = np.random.randint(0, max_defects + 1, size=N)
# Compute PMF (Probability Mass Function)
pmf_values = np.bincount(data) / N

# Compute CDF (Cumulative Distribution Function)
cdf_values = np.cumsum(pmf_values)

# Plot PMF
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.barplot(x=np.arange(len(pmf_values)), y=pmf_values, color="blue")
plt.xlabel("Number of Defective Items")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF)")

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(np.arange(len(cdf_values)), cdf_values, marker="o", linestyle="-", color="red")
plt.xlabel("Number of Defective Items")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF)")

plt.tight_layout()
plt.show()
```

```python
p_defect = 0.03  # 3% defect probability
N_bulbs = 100  # Number of bulbs checked

# (i) Probability of exactly 5 defective bulbs
prob_5_defective = stats.binom.pmf(5, N_bulbs, p_defect)

# (ii) Probability of at least 8 defective bulbs
prob_8_or_more = 1 - stats.binom.cdf(7, N_bulbs, p_defect)

# (iii) Expected number of defective bulbs
expected_defective = N_bulbs * p_defect

# Print results
print(f"(i) Probability of exactly 5 defective bulbs: {prob_5_defective:.4f}")
print(f"(ii) Probability of at least 8 defective bulbs: {prob_8_or_more:.4f}")
print(f"(iii) Expected number of defective bulbs: {expected_defective:.2f}")
```
